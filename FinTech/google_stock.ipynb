```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import matplotlib
import matplotlib.pyplot as plt

#retrieve the data from Google
goog = web.DataReader('GOOG', data_source='google',start='9/05/2011', end='02/21/2017')
goog.tail()
```

```python
#implement the necessary analytics for the volatilities
goog['Log_Ret'] = np.log(goog['Close'] / goog['Close'].shift(1));
goog['Volatility'] = pd.rolling_std(goog['Log_Ret'], window=252) * np.sqrt(252);

#plot the results
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
goog[['Close', 'Volatility']].plot(subplots=True, color='blue', figsize=(8,8));
```
